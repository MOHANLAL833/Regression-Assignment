```python
import pandas as pd
```

```python
data=pd.read_csv("insurance_pre.csv")
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
independent=data[["age", "bmi", "children", "sex_male","smoker_yes"]]
dependent=data[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
#Getting the parameters in the Dictionary
param_grid={'criterion':['squared_error','friedman_mse','absolute_error','poisson'],'splitter':['best','random'],'max_depth':[1,10,100,1000]}
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(xtrain,ytrain)
```

```python
re=grid.cv_results_
print ("The Rsquare value for the the best paramater {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("bmi:"))
children_input=float(input("children:"))
sex_input=int(input("Sex Male 0 or 1:"))
smoker_input=int(input("smoke yes 0 or 1:"))
```

```python
future_prediction=grid.predict([[age_input,bmi_input,children_input,sex_input,smoker_input]])
print("Future_prediction={}".format(future_prediction))
```
